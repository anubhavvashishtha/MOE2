```python
import pyautogui
import time

print("üü¢ Anti-sleep script started. Move your mouse to the bottom-right corner to stop.")

# Fail-safe is ON (moving to corner raises exception)
pyautogui.FAILSAFE = False  

# Get screen size
screen_width, screen_height = pyautogui.size()

try:
    while True:
        # Get current position
        x, y = pyautogui.position()
        
        # Check if in bottom-right corner (within 10px tolerance)
        if x > screen_width - 10 and y > screen_height - 10:
            print("üõë Mouse moved to bottom-right corner. Stopping...")
            break
        
        # Move slightly in a small pattern to keep awake
        pyautogui.moveRel(10, 0, duration=0.2)
        pyautogui.moveRel(-10, 0, duration=0.2)
        
        # Wait before next move
        time.sleep(10)

except KeyboardInterrupt:
    print("‚ö†Ô∏è Script interrupted manually.")
```
